### 1. Explique, com suas palavras, o que é machine learning?

Machine Learning é uma área da Inteligência Artificial focada no desenvolvimento de modelos que têm a capacidade de aprender padrões a partir de um conjunto de dados. É possível utilizar aprendizado supervisionado com o uso de um conjunto de dados rotulados, ou aprendizado não supervisionado quando não existe rotulação nos dados.

### 2. Explique o conceito de conjunto de treinamento, conjunto de validação e conjunto de teste em machine learning.

Em projetos de machine learning, os dados disponíveis costumam ser divididos em três grupos para garantir que o modelo aprenda corretamente e consiga generalizar para dados não vistos:

*   **Conjunto de Treinamento:** É a maior parte dos dados (geralmente 80%) e é usado para "ensinar" o modelo. O algoritmo analisa esses dados para identificar padrões e ajustar seus parâmetros internos (pesos).
*   **Conjunto de Validação:** É uma amostra de dados usada para avaliar o modelo frequentemente durante a fase de treinamento. Seu principal objetivo é ajudar a ajustar os hiperparâmetros.
*   **Conjunto de Teste:** É uma amostra separada de dados que nunca é mostrada ao modelo durante o treinamento ou a validação. Ele é usado apenas uma vez, no final do projeto, para avaliar o desempenho final do modelo. O conjunto de teste simula como o modelo irá se comportar no mundo real lidando com dados nunca vistos antes.

### 3. Explique como você lidaria com dados ausentes em um conjunto de dados de treinamento.

Lidar com dados ausentes é uma etapa crucial do pré-processamento. A estratégia escolhida depende da quantidade de dados que faltam e do tipo de variável. Algumas das abordagens mais comuns que eu utilizaria incluem:

1.  **Exclusão:** Se o conjunto de dados for muito grande e a porcentagem de dados ausentes for mínima (ou se uma coluna específica tiver a grande maioria dos valores ausentes), podemos simplesmente descartar as linhas ou colunas incompletas.
2.  **Imputação Estatística:** Consiste em substituir o valor ausente por uma medida estatística calculada sobre o restante dos dados daquela coluna. Podemos utilizar a **Média** ou **Mediana**  ou a **Moda**.
3.  **Imputação Preditiva:** Utilizar outros algoritmos de machine learning (como KNN, ou Árvores) para prever e preencher o valor ausente com base nos padrões das outras variáveis daquela mesma linha.

Antes de tudo é importante tentar identificar a origem da ausencia, como aprendido no bootcampt de ciencia de dados é possível atribuir classificações MAR MNAR MCAR para a ausência dos dados.

### 4. O que é uma matriz de confusão e como ela é usada para avaliar o desempenho de um modelo preditivo?

Uma matriz de confusão é uma tabela que cruza os resultados reais de um conjunto de dados com os resultados previstos pelo modelo, permitindo visualizar detalhadamente o desempenho de um modelo de **classificação**. 

Para um problema binário (duas classes, Positivo e Negativo), ela se divide em quatro quadrantes: VP, VN, FP, FN

Ela nos permite obter algumas métricas específicas para entender exatamente onde o modelo está errando. A partir dela calculamos a **Precisão** e o **Recall**.
